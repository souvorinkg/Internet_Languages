```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from io import StringIO


df = pd.read_csv('LangData.csv')
```

```python
plt.figure(figsize=(8, 6))

# Create a box plot for the variable 'C_PROD'
sns.boxplot(x=df['C_PROD'])

# Set the title and labels
plt.title('Spread of Content Production')
plt.xlabel('Content Production')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size (adjust as needed)
ax =sns.violinplot(x='EU', y='C_PROD', data=df)
plt.title('Impact of the EU')
ax.set(xlabel='Is it an EU language?', ylabel='Content Production')

plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size (adjust as needed)
ax =sns.violinplot(x='Col_Lang', y='C_PROD', data=df)
plt.title('Impact of Colonial Language')
ax.set(xlabel='Has the language been used in Colonial administration?', ylabel='Content Production')

plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size (adjust as needed)
ax =sns.violinplot(x='Top_Wiki', y='C_PROD', data=df)
plt.title('Impact of Wikipedia')
ax.set(xlabel='Language is the most used Wiki', ylabel='Content Production')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(x='Wiki_Score', y='C_PROD', data=df)

# Set the x-axis ticks explicitly
ax.set_xticks([0, 5, 10, 15, 20, 25, 30])

# Add regression line
sns.regplot(x='Wiki_Score', y='C_PROD', data=df, scatter=False, ax=ax, color='red')

plt.title('Impact of Wikipedia Size')
ax.set(xlabel='Size of Wikipedia', ylabel='Content Production')

plt.show()
```

```python
X = df[['EU', 'Col_Lang']]  # Independent variables
y = df['C_PROD']  # Dependent variable

# Add a constant to the independent variables (i.e., add an intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Get the regression summary
summary = model.summary()

print(summary)
```

RQ: Does membership in the EU cause high internet functionality for language speakers? 

Internet Functionality is a range of .8 between .65 and 1.45, with an average internet functionality score of .96. The skew of the data to languges below 1 can be attributed to the fact that there are a handful of languages very well represented on the internet, accounting for a long tail stretching to 1.45. However, the median language within Europe has an internet functionality of .93. This reflects the divide between the high status of the major languages of the internet compared to the lower status of the majority of languages. The IQR is 1.05 to .87, which means that a second shorter tail accounts for the lower end of the languages, with 76 % of the data falling within one standard deviation of the mean. 

The Languages of the EU are noted in Figure 1. Languages not included in the EU include Norwegian, Regional Languages of Italy and Spain, and the Netherlands, and many languages spread throughout Eastern Europe. 

In bivariate regression, EU membership, as well as Colonial Language, HDI, and Wiki_Score were all found to be significant. However, Membership in the EU only accounted for 26.6% of the variation in my internet functionality. 

Within the Multivariate Analysis shown in figure 2, the HDI and Wiki Score of language communities becomes highly significant, as well as whether the chosen language is a colonial language. However, whether a language is within the EU loses explanator power. Colonial Language is a binary indicator shown in Figure 3. Languages that have been used in a colonial context scored .08 points higher in internet functionality, equivilent to the jump from Chuvash to Danish, or from Finnish to French. 

The Wikiscore represents equivilent distances between the quality between the English Wikipedia, compared to the 328 other current Wikipedias. The difference between English(1st) and Swedish(11th), Swedish and Tatar(33rd), and Tatar and Kabardian(44th, lowest in dataset) is 10 deviations. To illustrate the relative differences, active users is one component of the Wiki_Score, with English having 122,000, Swedish having 2,100, Tatar having 81, and Kabardian having 16. This relationhship is both strong and leads to large changes in Internet Functionality, as decreasing your Wiki_Score by 1 (moving towards the English Wiki) leads to almost a full .01 increase in Internet Functionality. This makes sense, as languages with corpi have a mutual relationship with content, as corpi helps cite sources for the rest of the web, which in turn provides more sources for the corpus. 

HDI was included to account for economic factors not accounted for within the linguistic school. HDI for all populations within Europe is uniformly above .75, excluding the highly ostracized Romani people. Figure 5 shows the continuum between Western and Eastern Europe. HDI appears to have high predictive power in several cases, as the National Standards of Albanian or Macedonian have lower internet Functionality than comunities langages with an HDI of .05 higher, such as Hungarian or Romanian speakers. An increase in HDI by .05 is associated with a jump in .03 for Internet Functionality. Interestingly, Regional Dutch and Northern Italian Languages have some of the highest HDI among populations in the world, but all have some of the lowest Internet Functionality in the study.
